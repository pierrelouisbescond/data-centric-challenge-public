```python
import numpy as np
```

#### Let's create potential datasets sizes and errors numbers as NumPy arrays

```python
dataset_sizes = np.linspace(2000, 3000, 101, dtype="int")
print(dataset_sizes[0:10])

erroneous_labels = np.linspace(0, 1000, 1001, dtype="int")
print(erroneous_labels[0:10])
```

#### Let's iterate on each dataset size and error number to see any match with leaderboard scores

```python
target_scores = [0.86570, 0.85826, 0.85744, 0.85620, 0.85496, 0.85455, 0.85372, 0.85331, 0.85248, 0.85083]
dataset_size_memory = 0

for dataset_size in dataset_sizes:

    for erroneous_label in erroneous_labels:

        score_raw = (dataset_size-erroneous_label)/dataset_size
        score = round(score_raw, 5)

        if score in target_scores:

            if dataset_size_memory != dataset_size:

                print ("*"*20)
                dataset_size_memory = dataset_size

            print(str(dataset_size).ljust(4), ":", str(erroneous_label).ljust(3), ":", str(score).ljust(8) ,":", score_raw)
```

#### Let's infer the erroneous pictures number from the score

```python
dataset_size = 2420
```

```python
target_score = float(input("What is your current score?"))

erroneous_label = dataset_size

while erroneous_label > 0:

    if target_score == round((dataset_size-erroneous_label)/dataset_size, 5):

        print(f"Your model mislabelled {erroneous_label} pictures out of {dataset_size}!")
        break

    erroneous_label -= 1
```

#### Let's display full scores table of a specific dataset size

```python
dataset_size = 2440

for erroneous_label in erroneous_labels:

    score = round((dataset_size-erroneous_label)/dataset_size, 5)

    print(str(dataset_size).ljust(4), ":", str(erroneous_label).ljust(3), ":", score)
```
